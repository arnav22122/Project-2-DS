```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# %% [code]
print("="*80)
print("Project #2 - Data Analysis with Pandas".center(80))
print("Arnav Patel and Abhijay Madanagopal".center(80))
print("DS 220".center(80))
print("05/05/2025".center(80))
print("="*80)
print("In this project, I explore NBA player and team statistics from the 2024 season")
print("to uncover meaningful insights. The dataset was chosen because the NBA is a rich")
print("and complex domain with broad appeal, and analyzing player performance, team")
print("dynamics, and outcomes can offer useful insights to fans, analysts, and decision-makers.\n")

print("Six key questions we aim to address:\n")
print("1) Who are the top scorers in the 2024 NBA season, and how do they compare in assists and rebounds?")
print("2) What is the overall statistical profile of NBA players this season (age, minutes, field goals, etc.)?")
print("3) What are the strongest correlations between player performance metrics?")
print("4) How do team points per game relate to their win totals?")
print("5) How are players distributed across performance tiers (Elite, Star, Good, Average) based on PPG?")
print("6) What visual trends can we observe in team and player performance that could inform coaching or management?\n")


sns.set_theme(context='notebook', style='whitegrid', palette='deep', font_scale=1.0)
plt.rcParams['figure.dpi'] = 100

url_players = "https://www.basketball-reference.com/leagues/NBA_2024_per_game.html"
player_stats = pd.read_html(url_players, header=0)[0]

player_stats = player_stats[player_stats.Rk != 'Rk'].drop(columns='Rk')

player_stats = player_stats.apply(pd.to_numeric, errors='ignore')

url_teams = "https://www.basketball-reference.com/leagues/NBA_2024.html"
team_stats = pd.read_html(url_teams, header=0)[0]
team_stats = team_stats.loc[~team_stats['Eastern Conference'].isna()]

print(f"‚úÖ Loaded {len(player_stats)} players, {len(team_stats)} teams")

display(player_stats.describe().round(2))

top_pts = player_stats.nlargest(10, 'PTS')[['Player','PTS','AST','TRB']]
print("üèÄ Top 10 Scorers")
display(top_pts)


top_ast = player_stats.nlargest(10, 'AST')[['Player','AST','PTS','TRB']]
print("üÖ∞Ô∏è Top 10 Playmakers")
display(top_ast)

top_trb = player_stats.nlargest(10, 'TRB')[['Player','TRB','PTS','AST']]
print("üì¶ Top 10 Rebounders")
display(top_trb)

fig, axes = plt.subplots(1, 3, figsize=(14,4))

sns.histplot(player_stats.PTS, bins=30, kde=True, ax=axes[0])
axes[0].set(title='Points per Game', xlabel='PTS')
axes[0].text(0.95, 0.85, "Scoring spread", transform=axes[0].transAxes,
             ha='right', fontsize=9, color='gray')

sns.histplot(player_stats.AST, bins=30, kde=True, ax=axes[1])
axes[1].set(title='Assists per Game', xlabel='AST')
axes[1].text(0.95, 0.85, "Passing range", transform=axes[1].transAxes,
             ha='right', fontsize=9, color='gray')

sns.histplot(player_stats.TRB, bins=30, kde=True, ax=axes[2])
axes[2].set(title='Rebounds per Game', xlabel='TRB')
axes[2].text(0.95, 0.85, "Rebounding spread", transform=axes[2].transAxes,
             ha='right', fontsize=9, color='gray')

plt.tight_layout()
plt.show()


corr = player_stats.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10,8))
sns.heatmap(
    corr, mask=mask, cmap='vlag', vmin=-1, vmax=1,
    annot=True, fmt=".2f",
    annot_kws={"size":6},    
    cbar_kws={"shrink":0.5}
)
plt.xticks(fontsize=6, rotation=45, ha='right')
plt.yticks(fontsize=6)
plt.title("Correlation Matrix of Player Stats", pad=12)
plt.text(0.5, -0.1, "Higher red = strong positive relationship; blue = negative", 
         ha='center', va='top', transform=plt.gca().transAxes, fontsize=9, color='gray')
plt.tight_layout()
plt.show()


if 'PTS/G' in team_stats.columns and 'W' in team_stats.columns:
    plt.figure(figsize=(8,6))
    sns.regplot(
        x='PTS/G', y='W', data=team_stats,
        scatter_kws={'s':60,'alpha':0.7}, line_kws={'color':'crimson'}
    )
    plt.title("Team: Points/Game vs. Wins")
    plt.xlabel("Points per Game (PTS/G)")
    plt.ylabel("Wins (W)")
    plt.grid(True)
    plt.text(0.05, 0.95, "Red line = best-fit regression", transform=plt.gca().transAxes,
             ha='left', va='top', fontsize=9, color='gray')
    plt.tight_layout()
    plt.show()


bins = [0,10,18,25, np.inf]
labels = ['Average','Good','Star','Elite']
player_stats['Tier'] = pd.cut(player_stats.PTS, bins=bins, labels=labels, right=False)

counts = player_stats.Tier.value_counts().reindex(labels)
display(counts.to_frame('Count'))

plt.figure(figsize=(6,4))
sns.barplot(x=counts.index, y=counts.values, palette='muted')
plt.title("Player Performance Tiers")
plt.xlabel("Tier")
plt.ylabel("Number of Players")
plt.text(0.95, 0.85, "Elite ‚âà 2.5% of players", transform=plt.gca().transAxes,
         ha='right', fontsize=9, color='gray')
plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

sns.scatterplot(x='Age', y='MP', data=player_stats, ax=ax1, alpha=0.5)
ax1.set(title='Age vs Minutes', xlabel='Age', ylabel='Minutes/Game')
ax1.text(0.05, 0.95, "Veterans vs rookies", transform=ax1.transAxes,
         ha='left', va='top', fontsize=9, color='gray')

sns.regplot(x='Age', y='PTS', data=player_stats, ax=ax2,
            scatter_kws={'alpha':0.4}, line_kws={'color':'forestgreen'})
ax2.set(title='Age vs Points/Game', xlabel='Age', ylabel='PTS')
ax2.text(0.05, 0.95, "Green line = regression", transform=ax2.transAxes,
         ha='left', va='top', fontsize=9, color='gray')

plt.tight_layout()
plt.show()

 
# %% [code]
elite_count = len(player_stats[player_stats.PTS >= 25])
total_players = len(player_stats)
elite_percentage = 100 * elite_count / total_players

print("\n" + "="*80)
print("FINAL ANALYSIS & CONCLUSION".center(80))
print("="*80)
print("1) Top Scorers:")
print("   ‚Ä¢ Stars like Joel Embiid, Luka Donƒçiƒá, and Giannis Antetokounmpo not only led in points")
print("     but also contributed significantly in assists and rebounds, showing their all-around impact.\n")

print("2) Player Profile:")
print("   ‚Ä¢ The average NBA player is 26 years old, plays ~18 minutes per game, and scores ~8 points,")
print("     giving context to evaluate standout performers.\n")

print("3) Strongest Correlations:")
print("   ‚Ä¢ Field goals attempted (FGA) and points (PTS) are tightly linked, as are assists (AST) and points,")
print("     helping identify which metrics matter most.\n")

print("4) Team Points vs. Wins:")
print("   ‚Ä¢ There‚Äôs a clear positive relationship between team points per game and win totals,")
print("     though defensive outliers remind us that offense alone isn‚Äôt everything.\n")

print("5) Performance Tiers:")
print("   ‚Ä¢ Only {:,} of {:,} players (~{:.1f}%) averaged 25+ PPG, confirming the rarity of elite scorers."
      .format(elite_count, total_players, elite_percentage))
print("     Most players fall into the Average or Good tiers.\n")

print("6) Visual Trends:")
print("   ‚Ä¢ Heatmaps and count plots effectively communicate key patterns in player and team performance,")
print("     offering insights that can inform coaching and management decisions.\n")

print("Overall, this analysis shows how data-driven insights can benefit fans and professionals alike,")
print("highlighting both talent gaps and pathways to success.\n")
print("="*80)


```
